```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, log2
```

## Primero la función densidad

```python
def f_densidad(x):
    if x >= 1 and x < 2:
        return 1/9 * x - 1/9
    elif x >= 2 and x < 5:
        return 1/9
    elif x >= 5 and x < 7:
        return 2/45 * x - 1/9 
    elif x >= 7 and x <= 10:
        return -1/15 * x + 2/3
    elif x < 1 or x > 10:
        return 0
    else:
        print("No implemmentado para {0}".format(x))
        raise NotImplementedError
```

```python
# 100 linearly spaced numbers
x = np.linspace(0 ,11, 100)

# the function, which is y = x^2 here
y = [f_densidad(i) for i in x]

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x, y, color='#087E8B', linewidth=3.0)
plt.xlabel('x', size=15)
plt.ylabel('f(x)', size=15)
plt.show()
```

## Ahora, la función acumulativa

```python
def F_acumulativa(x):
    if x >= 1 and x < 2:
        return 1/18 * x**2 - 1/9 * x + 1/18
    elif x >= 2 and x < 5:
        return 1/9* x -1/6
    elif x >= 5 and x < 7:
        return 1/45 * x**2 - 1/9 * x + 7/18
    elif x >= 7 and x <= 10:
        return -1/30 * x**2 + 2/3 * x - 7/3
    elif x < 1:
        return 0
    elif x > 10:
        return 1
    else:
        print("No implemmentado para {0}".format(x))
        raise NotImplementedError
```

```python
# 100 linearly spaced numbers
x = np.linspace(0 ,11, 100)

# the function, which is y = x^2 here
y = [F_acumulativa(i) for i in x]

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x, y, color='#087E8B', linewidth=3.0)
plt.xlabel('x', size=15)
plt.ylabel('f(x)', size=15)
plt.show()
```

## Por último la función inversa

```python
def F_inversa_acumulativa(x):
    if x >= 0 and x < 1/18:
        return 1 + sqrt(18*x)
    elif x >= 1/18 and x < 7/18:
        return 9*x + 9/6
    elif x >= 7/18 and x < 7/10:
        return 15/6 + sqrt(45*x - 45/4) 
    elif x >= 7/10 and x <= 1:
        return 30/3 - sqrt(30 - 30*x)
    else:
        print("No implemmentado para {0}".format(x))
        raise NotImplementedError
```

```python
# 100 linearly spaced numbers
x = np.linspace(0 , 1, 10)

# the function, which is y = x^2 here
y = [F_inversa_acumulativa(i) for i in x]

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x, y, color='#087E8B', linewidth=3.0)
plt.xlabel('x', size=15)
plt.ylabel('f(x)', size=15)
plt.show()
```
