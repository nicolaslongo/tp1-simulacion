```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definimos el Generador Congruente Lineal utilizando los parámetros pedidos

```python
m = 2e32
a = 1013904223
c = 1664525
def GCL(Un):
    return (a*Un + c) % m
```

```python
u0 = np.array([89992, 90251, 98271, 98577]).mean()
u0
```

```python
## Generamos u1 a modo de prueba
u1 = GCL(u0)
u1
```

```python
def con_f_genero_N(f, N, seed):
    v = np.empty(N)
    v[0] = seed
    for i in range(1,N):
        v[i] = f(v[i-1])
    return v
```

```python
N = 2000
x = con_f_genero_N(GCL, N, u0)
print(x)

plt.figure(figsize=(12, 8), dpi=80)
plt.hist(x, bins = 100)
plt.show()
```

### Ahora modificaremos el GCL para que los valores que devuelva estén en el intervalo [0, 1]. Para esto, como es un generador lineal

```python
def GCL_normalizado(Un):
    return GCL(Un) / (m+1)
```

```python
# Generamos u1 a modo de prueba
u1 = GCL_normalizado(u0)
u1
```

```python
# Creo que hay que encontrar el T entonces para "normalizar" el resultado de GCL entre 0 y 1

N = 2000
x = con_f_genero_N(GCL_normalizado, N, u0)
print(x)

plt.figure(figsize=(12, 8), dpi=80)
plt.hist(x[1:], bins = 10)
plt.show()
```
